# Example: Quality Control and Manufacturing
## Scenario
Imagine a manufacturing company that produces metal rods for construction. The rods are supposed to have a diameter of 10 mm with a tolerance of ¬±0.1 mm. To ensure the quality of the rods, the company uses a normal distribution to model the diameter measurements and employs control charts and Six Sigma methodologies to detect deviations from the desired specifications.

Measure the diameter of a sample of rods at regular intervals. Suppose the sample size is 100 rods, and the measurements (in mm) are recorded.

### Create a Control Chart:

Plot the sample measurements on a control chart with the UCL and LCL to monitor the process over time.

Calculate the mean and standard deviation:


```python
import numpy as np

# Sample measurements
data = np.array([10.05, 9.97, 10.02, 10.03, 9.96, 10.04, 9.98, 10.01, 10.09, 9.99])

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

mean, std_dev
```

```python
import matplotlib.pyplot as plt

# Sample indices
samples = np.arange(1, len(data) + 1)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(samples, data, marker='o', linestyle='-', color='b', label='Diameter Measurements')
plt.axhline(y=mean, color='g', linestyle='-', label='Mean')
plt.axhline(y=10.1, color='r', linestyle='--', label='Specification Limit (Upper)')
plt.axhline(y=9.9, color='r', linestyle='--', label='Specification Limit (Lower)')
plt.axhline(y=10.089, color='orange', linestyle='--', label='Upper Control Limit (UCL)')
plt.axhline(y=9.921, color='orange', linestyle='--', label='Lower Control Limit (LCL)')
plt.xlabel('Sample Number')
plt.ylabel('Diameter (mm)')
plt.title('Control Chart for Rod Diameter')
plt.legend()
plt.show()
```

# Interpretation

* If all the points are within the control limits (UCL and LCL), the process is considered to be in control.

*  If any points fall outside the control limits, it indicates that there may be an issue with the manufacturing process that needs to be investigated and corrected. For example, the machinery might need recalibration, or there could be a problem with the raw materials.

*  Regularly monitoring the control chart helps in maintaining the quality of the rods and ensuring they meet the specified tolerances.

By using the normal distribution and control charts, the manufacturing company can ensure that the rods are consistently produced within the desired specifications, reducing the risk of defects and maintaining high-quality standards.
